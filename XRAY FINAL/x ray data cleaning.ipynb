```python
# Get data for positive covid samples
FILE_PATH = "Datasets/metadata.csv"
IMAGES_PATH = "Datasets/covid/images"
```

```python
covid_data = pd.read_csv(FILE_PATH)
print(covid_data.shape)
```

```python
covid_data.head()
```

```python
TARGET_DIR = "Datasets/covid_data"
if not os.path.exists(TARGET_DIR):
    os.mkdir(TARGET_DIR)
    print("covid folder created")
```

```python
count = 0
for(i,row) in covid_data.iterrows():
    if (row["finding"]=="Pneumonia/Viral/COVID-19") and row["view"]=="PA": #PA means frontview of X-RAY
        filename = row["filename"]
        image_path = os.path.join(IMAGES_PATH,filename)
        image_copy_path = os.path.join(TARGET_DIR, filename)
        shutil.copy2(image_path,image_copy_path)
        print("moving image",count)
        count+=1
        
print(count)
```

## Separating into train and test dir by creating a new folder as followed below

```python
COVID_FILE_PATH = "Datasets/covid_data"
TARGET_COVID_TRAIN_DIR = "Datasets/final_data/train/covid"
TARGET_COVID_TEST_DIR = "Datasets/final_data/test/covid"

HEALTHY_FILE_PATH = "Datasets/healthy_data"
TARGET_HEALTHY_TRAIN_DIR = "Datasets/final_data/train/normal" 
TARGET_HEALTHY_TEST_DIR = "Datasets/final_data/test/normal"
```

```python
covid_image_names = os.listdir(COVID_FILE_PATH)
healthy_image_names = os.listdir(HEALTHY_FILE_PATH)

#GETTING TEST DATA
for i in range(39):
    #COVID TEST DATA
    covid_image_name = covid_image_names[i]
    covid_image_path = os.path.join(COVID_FILE_PATH, covid_image_name)
    
    covid_test_target_path = os.path.join(TARGET_COVID_TEST_DIR, covid_image_name)
    
    shutil.copy2(covid_image_path, covid_test_target_path)
    
    #HEALTHY TEST DATA
    healthy_image_name = healthy_image_names[i]
    healthy_image_path = os.path.join(HEALTHY_FILE_PATH, healthy_image_name)
    
    healthy_test_target_path = os.path.join(TARGET_HEALTHY_TEST_DIR, healthy_image_name)
    
    shutil.copy2(healthy_image_path, healthy_test_target_path)
    
#GETTING TRAIN DATA
for i in range(39,196):
    #COVID TRAIN DATA
    covid_image_name = covid_image_names[i]
    covid_image_path = os.path.join(COVID_FILE_PATH, covid_image_name)
    
    covid_train_target_path = os.path.join(TARGET_COVID_TRAIN_DIR, covid_image_name)
    
    shutil.copy2(covid_image_path, covid_train_target_path)
    
    #HEALTHY TEST DATA
    healthy_image_name = healthy_image_names[i]
    healthy_image_path = os.path.join(HEALTHY_FILE_PATH, healthy_image_name)
    
    healthy_train_target_path = os.path.join(TARGET_HEALTHY_TRAIN_DIR, healthy_image_name)
    
    shutil.copy2(healthy_image_path, healthy_train_target_path)
    
```
