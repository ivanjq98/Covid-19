MOST RECENT STATE COVID HISTORY DATA

```python
res = requests.get('https://api.covidtracking.com/v1/states/daily.csv')
data = StringIO(res.text)
StateHistory = pd.read_csv(data)
```

```python
StateHistory = StateHistory[['state','date','death', 'positiveCasesViral', 'positive', 'positiveIncrease']]
```

```python
StatePolicy = pd.read_csv('datasets/StayAtHomeOrders.csv' )
StatePolicy = StatePolicy[['State_Tribe_Territory','Current_order_status','Effective_date','Expiration_date','date']]
```

```python
today = datetime.date.today()
t = datetime.time(0, 0)
Today = datetime.datetime.combine(today, t)
```

```python
StatePolicy['Start SHN'].replace({'Current_order_status': Today}, inplace = True)
StatePolicy['End SHN'].replace({'Effective_date': Today}, inplace = True)
```

```python
StateHistory = StateHistory.merge(StatePolicy, left_on = 'state', right_on='state')
```

```python
temp = pd.to_datetime(StateHistory['date'], format = '%Y%m%d')
StateHistory['date'] = temp
```

```python
result = (StateHistory.date > StateHistory['Start SHN']) & (StateHistory.date<=StateHistory['End SHN'])
StateHistory['on SHN'] = result
```

```python
StateHistory.to_csv('StateHistory.csv')
```
